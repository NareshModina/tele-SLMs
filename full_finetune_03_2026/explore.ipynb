```python
from datasets import load_dataset
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("HuggingFaceTB/SmolLM2-135M")

ds = load_dataset("HuggingFaceH4/ultrachat_200k", split="train_sft")
print(f"Examples: {len(ds):,}")

# Sample token count on first 1000 examples
total_tokens = 0
for example in ds.select(range(1000)):
    # Flatten all message content
    text = " ".join(m["content"] for m in example["messages"])
    total_tokens += len(tokenizer(text, add_special_tokens=False)["input_ids"])

avg_tokens = total_tokens / 1000
estimated_total = avg_tokens * len(ds)

print(f"Avg tokens per conversation : {avg_tokens:,.0f}")
print(f"Estimated total tokens      : {estimated_total/1e9:.2f}B")
```
